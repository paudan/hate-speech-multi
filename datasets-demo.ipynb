```python
import os
from numpy.random import randint
from datasets import DatasetDict
import pandas as pd
```

## Wide-type dataset (multitarget)

```python
dataset = DatasetDict.load_from_disk("data/hf_berkeley_multi_wide")
data = dataset['train']
pd.DataFrame(data.select(range(30)))
```

## Long-type dataset (multitask)

```python
dataset = DatasetDict.load_from_disk(os.path.join("data/hf_berkeley_multi", "target=age"))
data = dataset['train']
selected = randint(0, high=len(data), size=30)
pd.DataFrame(data.select(selected))
```

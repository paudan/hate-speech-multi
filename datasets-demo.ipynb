```python
import os
from numpy.random import randint
from datasets import DatasetDict
import pandas as pd
```

## Wide-type dataset (multitarget)

```python
dataset = DatasetDict.load_from_disk("data/hf_berkeley_multi_wide")
data = dataset['train']
pd.DataFrame(data.select(range(30)))
```

```python
data = data.to_pandas()
stat_columns = list(set(data.columns) - {'text', 'hatespeech', 'contains_hate'})
data[stat_columns].sum(axis=0).plot(kind='bar')
```

## Long-type dataset (multitask)

```python
dataset = DatasetDict.load_from_disk(os.path.join("data/berkeley_dataset_multi", "target=age"))
data = dataset['train']
selected = randint(0, high=len(data), size=30)
pd.DataFrame(data.select(selected))
```

```python
import os
from datetime import datetime
from multitask_trainer import train_eval_model
```

## E5+LoRa

```python
os.environ["CUDA_VISIBLE_DEVICES"]="0"

model_path = "intfloat/multilingual-e5-large"
cache_dir = '../cache'

def get_experiment_config(input_model_name):
    return [
        ("data/lith_dataset_countries_multi", f"outputs/lith_classifier_{input_model_name}_lora_countries_fc"),
        ("data/lith_dataset_race_multi", f"outputs/lith_classifier_{input_model_name}_lora_race_fc"),
        ("data/lith_dataset_political_multi", f"outputs/lith_classifier_{input_model_name}_lora_political_fc"),
        ("data/lith_dataset_orientation_multi", f"outputs/lith_classifier_{input_model_name}_lora_orientation_fc"),
        ("data/lith_dataset_balanced_countries_multi", f"outputs/lith_classifier_{input_model_name}_lora_balanced_countries_fc"),
        ("data/lith_dataset_balanced_race_multi", f"outputs/lith_classifier_lora_{input_model_name}_balanced_race_fc"),
        ("data/lith_dataset_balanced_political_multi", f"outputs/lith_classifier_{input_model_name}_lora_balanced_political_fc"),
        ("data/lith_dataset_balanced_orientation_multi", f"outputs/lith_classifier_{input_model_name}_lora_balanced_orientation_fc"),
    ]

exp_config = get_experiment_config("e5")
lora_args = {
    "init_lora_weights": "olora",
    "target_modules": "all-linear"
}
model_args = {
    'use_focal_loss': True
}
batch_size = 32
```

```python
print("Training started:", datetime.now())
```

```python
for data_dir, model_dir in exp_config:
    print(f"Training E5 classifier for data {data_dir}")
    _, results = train_eval_model(model_path, data_dir, cache_dir=cache_dir, save_model_dir=model_dir, 
                                  batch_size=batch_size, num_epochs=20, eval_batch_size=128, 
                                  use_lora=True, lora_args=lora_args, model_args=model_args)
    print(results)
```

```python
print("Training finished:", datetime.now())
```

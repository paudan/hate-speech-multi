```python
import os
from datetime import datetime
import pandas as pd
from multitask_trainer import train_eval_model
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.expand_frame_repr', False)
pd.set_option('display.precision', 3)

os.environ["CUDA_VISIBLE_DEVICES"]="0"
```

## E5+LoRa

```python
model_path = "intfloat/multilingual-e5-large"
cache_dir = '../cache'

def get_experiment_config(input_model_name):
    return [
        ("../augmentation/augmented_data/lith_dataset_augmented_countries", f"outputs/lith_classifier_{input_model_name}_lora_countries_augmented_fc"),
        ("../augmentation/augmented_data/lith_dataset_augmented_race", f"outputs/lith_classifier_{input_model_name}_lora_race_augmented_fc"),
        ("../augmentation/augmented_data/lith_dataset_augmented_political", f"outputs/lith_classifier_{input_model_name}_lora_political_augmented_fc"),
        ("../augmentation/augmented_data/lith_dataset_augmented_orientation", f"outputs/lith_classifier_{input_model_name}_lora_orientation_augmented_fc")
    ]

exp_config = get_experiment_config("e5")
lora_args = {
    "init_lora_weights": "olora",
    "target_modules": "all-linear"
}
model_args = {
    'use_focal_loss': True
}
batch_size = 32
```

```python
print("Training started:", datetime.now())
```

```python
for data_dir, model_dir in exp_config:
    print(f"Training E5 classifier for data {data_dir}")
    _, results = train_eval_model(model_path, data_dir, cache_dir=cache_dir, save_model_dir=model_dir, 
                                  output_dir="outputs-train/e5-augmented-fc",
                                  batch_size=batch_size, num_epochs=20, eval_batch_size=128, 
                                  use_lora=True, lora_args=lora_args, model_args=model_args)
    print(results)
```

```python
print("Training finished:", datetime.now())
```
